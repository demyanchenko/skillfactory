```python
# Дана переменная name, в которой хранится имя пользователя, записанное русскими буквами. Напишите программу, 
# которая перебирает имя по буквам и для каждой буквы выводит сообщение: "... - гласная буква" или 
# "... - согласная буква", где вместо многоточия нужно подставить букву из имени пользователя.
name = 'Севастиан'
consonant = ['б','в','г','д','ж','з','к','л','м','н','п','р','с','т','ф','х','ц','ч','ш','щ']
vowel = ['а','е','ё','и','о','у','ы','э','ю','я']
for letter in name:
    if letter.lower() in vowel:
        print(letter, '- гласная буква')
    elif letter.lower() in consonant:
        print(letter, '- согласная буква')
    else:
        print(letter, '- никакая буква')
```

```python
# Превращаем весь файл в систему вложенных списков
import re
p = re.compile('\d+')
exams = []
f = open('StudentsPerformance.csv')
for line in f:
    if line.split(',')[0][1:-1] == 'gender':
        continue
    new_line = []
    for item in line[0:-1].split(','):
        if p.search(item) != None:
            new_line.append(int(p.search(item)[0]))
        else:
            new_line.append(item[1:-1])
    exams.append(new_line)
f.close()
exams
```

```python
# Вычислите средний балл абитуриентов на экзамене по чтению (reading score).
reading_score_sum = 0
for student in exams:
    reading_score_sum += student[6]

reading_score_avg = reading_score_sum/len(exams)
reading_score_avg
```

```python
# Сколько абитуриентов получили на экзамене по чтению (reading score) оценку ниже среднего?
count_below_avg = 0
for student in exams:
    if student[6] < reading_score_avg:
        count_below_avg += 1
count_below_avg
```

```python
# Какой средний балл экзамене по чтению (reading score) получили девочки?
female_reading_score_sum = 0
female_count = 0
for student in exams:
    if student[0] == 'female':
        female_reading_score_sum += student[6]
        female_count += 1
female_reading_score_avg = female_reading_score_sum/female_count
female_reading_score_avg
```

```python
# Сколько абитуриентов получили на экзамене по письму (writing score) оценку выше 90?
writing_score_count = 0
for student in exams:
    if student[7] > 90:
        writing_score_count += 1
writing_score_count
```

```python
# Сколько процентов абитуриентов, получивших на экзамене по письму (writing score) оценку более 90, 
# хорошо пообедали перед экзаменом (lunch = standard)?
lunch_count = 0
for student in exams:
    if student[7] > 90 and student[3] == 'standard':
        lunch_count += 1
lunch_count/writing_score_count*100
```

```python
# ...и тут пошла жаришка!!!
students = {}
f = open('StudentsPerformance.csv')
for line in f:
#     info = []
#     for item in line.split(','):
#         info.append(item[1:-1])
    if line.split(',')[0][1:-1] == 'gender':
        continue
    ethnicity = line.split(',')[1][1:-1]
    if ethnicity in students:
        students[ethnicity] += 1        
    else:
        students[ethnicity] = 1
f.close()
students
```

```python
# Создаём вложенный словарь
students = {}
f = open('StudentsPerformance.csv')
for line in f:
#     info = []
#     for item in line.split(','):
#         info.append(item[1:-1])
    if line.split(',')[0][1:-1] == 'gender':
        continue
    ethnicity = line.split(',')[1][1:-1]
    parents = line.split(',')[2][1:-1]
    if ethnicity in students:
#         students[ethnicity] += 1        
        if parents in students[ethnicity]:
#             increment parent
            students[ethnicity][parents] += 1
        else:
            students[ethnicity][parents] = 1
    else:
        students[ethnicity] = {}
        students[ethnicity][parents] = 1
f.close()
students
```

```python
# Сколько мальчиков хорошо пообедали перед экзаменом (lunch = standard)?
students = {}
f = open('StudentsPerformance.csv')
for line in f:
    gender = line.split(',')[0][1:-1]
    lunch = line.split(',')[3][1:-1]
    if gender == "gender":
        continue
    if gender in students:
        if lunch in students[gender]:
            students[gender][lunch] += 1
        else:
            students[gender][lunch] = 1
    else:
        students[gender] = {}
        students[gender][lunch] = 1
f.close()
students
```

```python
# Сколько девочек, родители которых имеют степень магистра, набрали по математике больше 90 баллов?
students = {}
f = open('StudentsPerformance.csv')
for line in f:
    gender = line.split(',')[0][1:-1]
    parent_edu = line.split(',')[2][1:-1]
    if gender == "gender":
        continue
    math_score = int(line.split(',')[5][1:-1])
    if math_score > 90:
        if gender in students:
            if parent_edu in students[gender]:
                cur_parent = students[gender][parent_edu]
                students[gender][parent_edu].append(math_score)
            else:
                students[gender][parent_edu] = [math_score]
        else:
            students[gender] = {}
            students[gender][parent_edu] = [math_score]

f.close()
students
```

```python
# Какой средний балл, полученный мальчиками на экзамене по чтению? Ответ округлите до трёх цифр после запятой.
male_reading_score = []
score_sum = 0
f = open('StudentsPerformance.csv')
for line in f:
    if line.split(',')[0][1:-1] == 'gender':
        continue
    if line.split(',')[0][1:-1] == 'male':
        male_reading_score.append(int(line.split(',')[6][1:-1]))
        score_sum += male_reading_score[-1]
f.close()
score_sum/len(male_reading_score)
```

```python
# Какой средний балл на экзамене по чтению набрали ученики, набравшие максимальный балл на экзамене по математике? 
score_sum = 0
stud_count = 0
f = open('StudentsPerformance.csv')
for line in f:
    gender = line.split(',')[0][1:-1]
    if gender == 'gender':
        continue
    math_score = int(line.split(',')[5][1:-1])
    reading_score = int(line.split(',')[6][1:-1])
    if math_score == 100:
        score_sum += reading_score
        stud_count += 1
f.close()
score_sum/stud_count
```

```python
# Какой средний балл на экзамене по письму набрали ученики, которые плохо пообедали 
# перед экзаменом (lunch = free/reduced)? Ответ округлите до двух цифр после запятой.
score_list = []
import re
p = re.compile('\d+')
f = open('StudentsPerformance.csv')
for line in f:
    if line.split(',')[0][1:-1] == 'gender':
        continue
    lunch = line.split(',')[3][1:-1]
    writing_score = p.search(line.split(',')[7][1:-1])[0]
    if lunch == 'free/reduced':
        score_list.append(writing_score)
f.close()
# Считаем среднее значение
score_sum = 0
for i in score_list:
    score_sum += int(i)
score_sum/len(score_list)
```
